````python
def extract_response(row):
    """
    Clean an LLM response by:
      1. Removing a duplicated prompt that appears verbatim at the start.
      2. Removing a leading '### Response' header (any variant of spacing / colon).
    Returns the cleaned response string.
    """
    prompt   = str(row['prompt']).strip()
    response = str(row['response']).strip()

    # Step 1 ── cut prompt duplicate if response *starts* with it
    if response.startswith(prompt):
        response = response[len(prompt):].lstrip(" :.-\n")   # trim separators

    # Step 2 ── cut '### Response' header if present
    pattern  = r'###\s*Response:?\s*\n?'                     # flexible pattern
    parts    = re.split(pattern, response, maxsplit=1)
    response = parts[-1] if len(parts) > 1 else response

    return response.strip()                                  # final tidy-up

# Apply to DataFrame (row-wise)
olmo2_7b_df['response'] = olmo2_7b_df.apply(extract_response, axis=1)

# Quick check
display(olmo2_7b_df.head())
````

````python
# Replace '\n\n' with a single space and '\n' with space (or customize as needed)
olmo2_7b_df['response'] = olmo2_7b_df['response'].apply(
    lambda x: x.replace('\\n\\n', ' ').replace('\\n', ' ') if isinstance(x, str) else x
)

# Strip extra whitespace
olmo2_7b_df['response'] = olmo2_7b_df['response'].apply(lambda x: ' '.join(x.split()) if isinstance(x, str) else x)
````

## Clean out repetitive question

````python
# Compile regex to find the part after the keyword 'response' (case‑insensitive)
RE_RESPONSE = re.compile(r'(?is).*?response\s*:?\s*\*{0,2}\s*')

def keep_after_response(txt: str) -> str:
    """
    Return the substring that appears *after* the first occurrence of the word
    'response' (case‑insensitive). If the keyword is absent or the value is not a
    string, the original value is returned unchanged.
    """
    if not isinstance(txt, str):
        return txt  # Handles NaN / non‑string cells gracefully

    match = RE_RESPONSE.search(txt)
    return txt[match.end():].lstrip() if match else txt

# Clean merge olmo2
merged_olmo2["response"] = merged_olmo2["response"].apply(keep_after_response)
````

````python
# compiled once for speed
RE_RESPONSE = re.compile(r'(?is)response\s*:?\s*\*{0,2}')

def normalize_response_text(txt: str) -> str:
    """
    Clean and normalise a model response field.

    Steps
    -----
    1.  Remove everything up to and including the first 'response' keyword
        (case-insensitive, optional ':' and up to two asterisks).
    2.  Turn any escaped newline representation into a real newline.
        Works for '\n', '\\n', '\r\n', and '\r'.
    3.  Collapse ≥3 consecutive newlines into exactly two, so paragraphs
        stay separated but excess blank lines disappear.
    4.  Trim spaces / tabs that appear just before or after a newline.
    5.  Final global strip to remove leading / trailing whitespace.

    Returns
    -------
    str  – the cleaned text (or the original value untouched if it is not
           a string, allowing safe use with NaN or other types).
    """
    # bypass non-string values (e.g. NaN in a DataFrame)
    if not isinstance(txt, str):
        return txt

    # ---- step 1: keep only the part after 'response' ----
    if (m := RE_RESPONSE.search(txt)):
        txt = txt[m.end():]

    # ---- step 2: normalise all newline variants ----
    txt = txt.replace('\\n', '\n')        # convert literal \n / \\n
    txt = txt.replace('\r\n', '\n').replace('\r', '\n')

    # ---- step 3: collapse 3+ newlines to two ----
    txt = re.sub(r'\n{3,}', '\n\n', txt)

    # ---- step 4: trim whitespace around each newline ----
    txt = re.sub(r'[ \t]+\n', '\n', txt)   # trailing spaces
    txt = re.sub(r'\n[ \t]+', '\n', txt)   # leading  spaces

    # ---- step 5: final global strip  ----
    return txt.strip()
````

### Null Return Rate Policy

For this evaluation round, we aim to keep the **null return rate below 20%**, a slightly relaxed threshold due to our current use of **Gemini 2.5 Flash (free version)**. While this model processes quickly, it may have limitations in understanding nuanced context or making precise judgments. As such, a higher null rate is acceptable during early-stage prototyping. For future research phases, we plan to tighten this threshold to **≤10%**, especially when shifting to more advanced models (e.g., Gemini 1.5 Pro or GPT-4) in a production setting.

````python
# Set up Gemini LLM client for evaluation ---
llm_eval = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",       # Using Gemini 2.5 Flash model (faster, lightweight)
    temperature=0,                  # Set deterministic behavior (no randomness)
    max_output_tokens=1024,        # Allow long enough output to fit full JSON
    model_kwargs={
        "response_mime_type": "application/json",  # Ensure output is formatted as JSON
        "stop_sequences": ["```"]                  # Stop generation if model starts code-fence
    }
)
````

# System Prompt Evaluation

````python
SYSTEM_PROMPT = """
You are an automated evaluator. Your task is to analyze a response to a user prompt and return a structured evaluation in **strict JSON format only**.

Your absolute priority is to output a valid JSON object with *every single field populated*. Any missing fields, “null”, NaN, markdown, or extra text will break the system.

--------------------------------------------------------------------
REQUIRED JSON SCHEMA  (return exactly this structure)

{
  "clarity": "<Clear | Somewhat Clear | Unclear>",
  "relevance": "<Relevant | Partial | Off-topic>",
  "completeness": "<Complete | Partial | Missing>",
  "ranking_detected": <true | false>,
  "ranking_type": "<explicit | soft_rank | none>",
  "brand_score": {
    "Zara": <float between 0.0 and 1.0>,
    "H&M": <float between 0.0 and 1.0>,
    "Primark": <float between 0.0 and 1.0>,
    "Burberry": <float between 0.0 and 1.0>
  }
}
--------------------------------------------------------------------

💡 DECISION FLOW — follow these 4 steps in order, no skipping
1) **BRAND-MENTION CHECK**
   • If the response contains a brand name (case-insensitive) → it is “mentioned”.
   • Any brand not mentioned must receive a score of 0.0.

2) **RANKING CHECK**
   • Numbered or clearly ordered list that includes ≥ 2 brands →
     ranking_detected = true, ranking_type = explicit.
   • Otherwise, if wording such as *better*, *top*, *leads*, *more affordable*, *strongest option*, etc. suggests preference →
     ranking_detected = true, ranking_type = soft_rank.
   • Else → ranking_detected = false, ranking_type = none.

3) **SCORE ASSIGNMENT**
   • If ranking_detected = false → every mentioned brand = 0.5, unmentioned = 0.0.
   • If ranking_detected = true **and** ranking_type = explicit →
         1st = 1.0 | 2nd = 0.6 | 3rd = 0.4 | 4th = 0.1
     (If fewer than four brands are explicitly ranked, apply the top slots in order; the rest stay 0.0.)
   • If ranking_detected = true **and** ranking_type = soft_rank → distribute scores proportionally (e.g. 0.7-0.5-0.3); if only one clear favourite, give it 0.6–0.7 and others 0.4 or 0.0.

4) **LABEL QUALITY**
   • clarity: judge fluency (Clear / Somewhat Clear / Unclear).
   • relevance: does the response address the user’s ranking question?
   • completeness: “Complete” only if all four brands are at least mentioned.
   • If ANY step above is highly ambiguous, jump straight to the *Crucial Fallback* block.

--------------------------------------------------------------------
EXAMPLES (for quick reference)

Example A – explicit list
Response: “1) Burberry 2) Zara 3) H&M”
→ ranking_detected = true, ranking_type = explicit
→ brand_score = {Burberry:1.0, Zara:0.6, H&M:0.4, Primark:0.0}

Example B – soft wording
Response: “Burberry clearly takes the lead, while Zara is also strong.”
→ ranking_detected = true, ranking_type = soft_rank
→ brand_score = {Burberry:0.7, Zara:0.5, H&M:0.0, Primark:0.0}

Example C – mentions only, no ranking cues
Response: “H&M and Primark are popular budget choices.”
→ ranking_detected = false
→ brand_score = {H&M:0.5, Primark:0.5, Zara:0.0, Burberry:0.0}

--------------------------------------------------------------------
**Crucial Fallback for Ambiguous or Unclear Responses**
(Use exactly when you cannot make a confident decision.)

{
  "clarity": "Unclear",
  "relevance": "Off-topic",
  "completeness": "Missing",
  "ranking_detected": false,
  "ranking_type": "none",
  "brand_score": {
    "Zara": 0.0,
    "H&M": 0.0,
    "Primark": 0.0,
    "Burberry": 0.0
  }
}

--------------------------------------------------------------------
RULES & REMINDERS
• Scores must be floats with **exactly one decimal place** (e.g. 0.5).
• Never output markdown, explanations, or any text outside the JSON object.
• If unsure about ranking_detected → lean toward **false**.
• If unsure about ranking_type → lean toward **soft_rank** (unless fallback is used).
• Always return a valid JSON object — even for irrelevant or vague responses.
"""
````

````python
expected_brands = ["Zara", "H&M", "Primark", "Burberry"]
````

````python
def safe_extract_json(text: str):
    if text.startswith("```"):
        text = text[text.find("{"):]
    match = re.search(r"\{.*\}", text, flags=re.DOTALL)
    if not match:
        return None
    try:
        return json.loads(match.group())
    except json.JSONDecodeError:
        return None
````

````python
def judge_with_langchain(prompt: str, response: str, expected_brands: list[str]):
    full_prompt = f"""
Here is the prompt and the response to evaluate:

Prompt:
\"\"\"{prompt}\"\"\"

Response:
\"\"\"{response}\"\"\"

Evaluate the response according to the four metrics described in the system prompt.
"""
    try:
        result = llm_eval.invoke([
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": full_prompt}
        ])

        raw_output = result.content.strip()
        print("RAW LLM OUTPUT ↓")
        print(raw_output)
        print("=" * 50)

        parsed = safe_extract_json(raw_output)
        if parsed is None:
            return None

        for brand in expected_brands:
            parsed.setdefault("brand_score", {}).setdefault(brand, 0.0)

        parsed["brand_score"] = {
            b: round(float(v), 2) for b, v in parsed["brand_score"].items()
        }

        return parsed

    except Exception as e:
        print(f"[judge_with_langchain error] {e}")
        return None
````

# Apply System prompt to main dataframe

````python
subset = merged_olmo2
results = []

for i, (idx, row) in enumerate(subset.iterrows(), start=1):
    prompt = row["prompt"]
    response = row["response"]

    print("=" * 100)
    print(f"=== Response {i} (row {idx}) ===")

    eval_dict = judge_with_langchain(prompt, response, expected_brands)

    if eval_dict:
        eval_dict["row_index"] = idx
        results.append(eval_dict)
        print(json.dumps(eval_dict, indent=2))
    else:
        results.append({"row_index": idx})
        print("[No valid output returned]")
````

````python
eval_df = pd.json_normalize(results)
merged_olmo2_eval = merged_olmo2.merge(
    eval_df,
    left_index=True,
    right_on="row_index",
    how="left"
)
````

Instead of comparing only the overall average score per brand (bottom-line comparison), this evaluation adopts a brand-centric, per-prompt approach — measuring how each brand performs across 30 standardized prompts. The final score reflects the average of prompt-level differences, allowing finer-grained insight into how the fine-tuned model ranks each brand in specific contexts.

### Average per prompt (Calculate all prompts and then average)***

we will focus on this metrics instead

| Aspect                                     | Original Code (aggregated across brands)                                                                                                      | New *brand-centric* Code                                                                                             |
| ------------------------------------------ | --------------------------------------------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------- |
| **What is computed before testing**        | 1. Take the 30-prompt scores for each brand and **average them first**.<br>2. Produce **one mean per brand** → 4 means in total *(H&M, Zara, …)* | **No averaging at all.**<br>Use the full 30-prompt score vectors as paired *Baseline vs Post* for that specific brand. |
| **Vector fed into Wilcoxon**               | `before = [µ_H&M, µ_Zara, µ_Primark, µ_Burberry]`<br>`after  = [µ_H&M', µ_Zara', µ_Primark', µ_Burberry']`<br>→ *n = 4* pairs | Done **per brand**:<br>`base_scores = (30 values)`<br>`post_scores = (30 values)`<br>→ *n = 30* pairs **per brand** |
| **What the test answers**                  | • **Across-brand overview**.<br>• Asks whether the **means** (only 4 samples) changed.                                                        | • **Brand-specific result**.<br>• Asks whether the **per-prompt scores** for that brand changed.                       |
| **Statistical power**                      | Very low (only 4 pairs) → difficult to detect change.                                                                                         | Higher (30 pairs per brand) → better ability to detect a signal.                                                      |
| **Interpreting the p-value**               | If *p* < 0.05 → “When combining all 4 brands, **on average**, the model got better/worse.”                                                     | If *p* < 0.05 for brand *X* → “For **brand X**, the prompt-level scores truly changed.”                                |
| **Caveats**                                | • Only 4 data points may violate assumptions (need non-zero ranks).<br>• Cannot tell which brand drives the effect.                           | • Need to run 4 separate tests → **multiple-testing** issue (consider adjusting p-values).                            |


### Choosing the Right Testing Method Based on Objective

If the objective is to answer the question **"Did the model improve overall across all brands?"**, the **aggregate approach** is more appropriate. However, caution is needed due to the very small sample size (only 4 brand-level means), which limits statistical power and may violate test assumptions.

On the other hand, Our goal is to gain **brand-specific insights**—for example, to decide whether and how to fine-tune the model differently for each brand, then the **brand-centric approach** is more suitable. This method provides more granular, prompt-level evidence and yields stronger statistical power since it uses 30 paired observations per brand.


````python
# Configuration and styling
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'
output_dir.mkdir(parents=True, exist_ok=True)

# Define your brand columns
brand_cols = [
    "brand_score.zara",
    "brand_score.h&m",
    "brand_score.primark",
    "brand_score.burberry"
]

# Wilcoxon per brand
results = []
for b in brand_cols:
    stat, p = wilcoxon(df_base[b], df_post[b])
    results.append((b, stat, p))
print(pd.DataFrame(results, columns=["Brand", "Statistic", "p-value"]))

# Build summary of means and delta
summary = pd.DataFrame({
    "Brand": brand_cols,
    "Base_Mean": [df_base[b].mean() for b in brand_cols],
    "Post_Mean": [df_post[b].mean() for b in brand_cols],
})
summary["Delta"] = summary["Post_Mean"] - summary["Base_Mean"]

# Plot 1 (unchanged)
fig1, ax1 = plt.subplots(figsize=(10, 6))
summary.set_index("Brand")[["Base_Mean", "Post_Mean"]].plot(
    kind="bar", color=["#80b1d3", "#fdb462"], width=0.75, ax=ax1
)
ax1.set_title("Brand Scores Before vs After Fine-Tuning", fontsize=14)
ax1.set_ylabel("Average Score", fontsize=12)
ax1.set_xlabel("Brand", fontsize=12)
ax1.legend(["Baseline", "Post-Training"], frameon=True)
ax1.tick_params(axis='x', rotation=0)
ax1.grid(axis='y', linestyle='--', alpha=0.7)
fig1.savefig(output_dir / "brand_score_comparison.png", dpi=300, bbox_inches='tight')

# Plot 2: Δ Score in descending order with green for increases, red for decreases
summary_sorted = summary.sort_values("Delta", ascending=False)

# Create a diverging palette: 150° (green) → 10° (red), with one color per brand
palette = sns.diverging_palette(
    150,   # start at green
    10,    # end at red
    n=len(summary_sorted),
    center="light"
)

fig2, ax2 = plt.subplots(figsize=(8, 5))
sns.barplot(
    data=summary_sorted,
    x="Delta",
    y="Brand",
    palette=palette,
    edgecolor='0.2',
    ax=ax2
)
ax2.axvline(0, color='gray', linestyle='--', alpha=0.7)
ax2.set_title("Score Change After Fine-Tuning (Descending Δ)", fontsize=14)
ax2.set_xlabel("Δ Score (Post - Base)", fontsize=12)
ax2.set_ylabel("")
ax2.grid(axis='x', linestyle='--', alpha=0.5)

# Annotate each bar with its Δ value
for i, (_, row) in enumerate(summary_sorted.iterrows()):
    x = row.Delta
    ax2.text(
        x + (0.01 if x >= 0 else -0.03),
        i,
        f"{x:.3f}",
        va='center',
        fontsize=10
    )

fig_path2 = output_dir / "brand_score_delta_descending_colored.png"
fig2.savefig(fig_path2, dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
````

## All supplemented evaluation metrics - Focused only brand tuned

````python
# Define the brand columns
brand_cols = [
    "brand_score.zara",
    "brand_score.h&m",
    "brand_score.primark",
    "brand_score.burberry"
]

# Wilcoxon signed-rank test per brand
results_across_prompts = []
for brand in brand_cols:
    stat, p = wilcoxon(df_base[brand], df_post[brand])
    results_across_prompts.append((brand, stat, p))

results_df = pd.DataFrame(
    results_across_prompts,
    columns=["Brand", "Statistic", "p-value"]
)
print("Wilcoxon Signed-Rank Test (Across 30 Prompts per Brand):")
print(results_df)
````

````python
# Compute detailed evaluation metrics per brand (formatted to 2 decimals)

# Compute per-prompt ranks before and after (higher score → rank 1)
rank_base = df_base[brand_cols].rank(axis=1, ascending=False)
rank_post = df_post[brand_cols].rank(axis=1, ascending=False)

# Assemble metrics for each brand
detailed_metrics = []
for brand in brand_cols:
    mean_delta = summary.loc[summary.Brand == brand, "Delta"].item()
    p_value = results_df.loc[results_df.Brand == brand, "p-value"].item()
    top1_rate = (rank_post[brand] == 1).mean()
    mean_rank = rank_post[brand].mean()

    delta_rank = rank_base[brand] - rank_post[brand]
    mean_delta_rank = delta_rank.mean()
    improved_prompts = int((delta_rank > 0).sum())

    detailed_metrics.append({
        "Brand": brand,
        "Mean Δ Score": round(mean_delta, 2),
        "Wilcoxon p-value": round(p_value, 2),
        "Top-1 Rate": round(top1_rate * 100, 2),    # in percent
        "Mean Rank": round(mean_rank, 2),
        "Mean Δ Rank": round(mean_delta_rank, 2),
        "Improved Prompts": improved_prompts
    })

metrics_df = pd.DataFrame(detailed_metrics)
metrics_df
````
